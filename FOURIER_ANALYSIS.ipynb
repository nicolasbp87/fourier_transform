# Fourier Transform in Time Series Analysis

## 1. Introduction to Time Series Analysis

A time series is a sequence of data points collected or recorded at successive points in time, typically at uniform intervals. These data points can represent any variable that changes over time. This temporal ordering is the key characteristic that distiguishes time series from other data types, and necessitates specialized analysis techniques. 

Time series analysis involves using statistical methods to analyze time series data in order to extract meaningful statistics and characteristics of the data, identify patterns sush as trends and seasonal variations, and, often, to forecast future values based on historical patterns. This analysis is crucial in various fields, including economics, industry, telecommunications, healthcare, meteorology, neuroscience, physics, and other areas where understanding temporal dynamics is essential for decision making, scientific discovery and development.

Time series analysis presents several important challenges, largely due to the nature of time series data and the complexities involved in modeling temporal dynamics accurately. Some of these chalenges include:

1. **Seasonality and Trend Decomoposition**: Many time series exhibit trends and seasonal patterns that need to be identified and separated from noise. Distinguishing between these components and erratic fluctuations can be challenging, especially when the patterns are not straightforward or evolve over time.
<br>

2. **Stationarity**: Many models assume that the data in a time series is stationary, meaning its statistical properties such as mean, variance and aucorrelation are constant over time. However, most rea. worls data exhibit trends, seasonality, and other forms of non-stationarity, making it necessary to transform the data before anlysis.
<br>

3. **Forecasting Accuracy**: Predicting future values based on past observations is a primary goal of time series analysis. Achieving high forecasting metrics, especially for long term predictions, is challenging due to the influence of randomness, model overfitting, and the inherent uncertainty in future events.
<br>

4. **High Dimensionality**: Some time series datasets can involve multiple interrelated variables (*multivariate time series*) making the analysis significantly more complex. Modeling the relantionship and interactions between these variables can be computationally intensive and methodologically challenging.
<br>

5. **Missing Data**: Data often has gaps due to errors in data collection, transmission losses, or other disruptions. Handling missing data appropriately is crucial, as improper imputation or deletion can lead to biased analyses and misleading results.
<br>

6. **Noise and Outliers**: Real world data can be noisy and contain outliers due to rare events, measurement errors, or other anomalies. Distinguishing between noise and true signal, and dealing with outliers appropriately, is essential for accurate analysis and interpretation.
<br>

7. **Time Series Length**: Lenght may affect time series analysis. Short series may not provide enough data to identify patterns or build reliable models, while very long time series can be computationally challenging to analyze and may include structural changes over time that invalidate earliear patterns.
<br>

8. **Dynamic Relationships**: In many applications, the relationships between variables in a time series can change over time. Accounting for these dynamic interactions and ensuring that models adapt to such changes is an important challenge.
<br>

9. **Computational Complexity**: Advanced time series analysis techniques, especially those involving machine learning and deep learning models, can be computationally expensive, requiring significant resources for data processing, model training, and validation.

Domain expertise is crucial when addressing these challenges and in every step of the data processing, analysiss and interpretation pipeline.


## 2. Basics of the Fourier Transform

The Fourier Transform is named after the french mathematician and physicist Jean-Baptiste Joseph Fourier, and is rooted in his work on heat transfer and vibration. The Fourier Transform is a mathematical technique that transforms a function of time (a signal or a time series) into a function of frequency. It decomposes a complex signal into sinusoids (sine and cosine functions) of different frequencies, providing a way to analyze the frequency components within a signal. This transformation from the time domain to the frequency domain allows for a clearer understanding of the underlying behaviours of the signal, not immediately abailable in the time domain. These behaviours include features like the following:

* **Periodic Patterns**: In the time domain, it might be challenging to discern precise periodicities. The Fourier Transform can clearly identify these periodic patterns by revealing peaks at frequencies corresponding to the signals periodic components.
<br>

* **Dominant Frequencies**: Some signals contain dominat frequencies that greatly influence their behaviour that are hidden in the time domain, but identifiable with the Fourier transform.
<br>

* **Noise Characteristics**: Noise in a time series often appears as random fluctuations, making analyzing the underlying signal harder. By transforming the data to the frequency domain, it's possible to differentiate between the signal and noise. High frequency components, for example, often represent noise, and thus, can be filtered out.
<br>

* **Tren Analysis**: While trends may be observable in the time domain, the Fourier Transform can quantify the strength and significance of these trends by showing low frequency components. This is particularly usegul for signals where trends are not linear or are obscured by other components.
<br>

* **Seasonality and Cycles**: For time series data with seasonal variations or cycles, data transformation can highlight these cycle frequencies and amplitude, providing insights into seasonal effects and cyclic behaviour.
<br>

* **System Resonances**: This is where a system vibrates at an increased amplitude at certain frequencies, know as the systems *natural resonant frequencies*. Identifying these frequencies can help in designing systems that avoid these resonances.
<br>

* **Harmonics**: The Fourier Transform can identify integer multiples of a *fundamental frequency*. These multiples are known as *harmonics*, and can be a relevant feature in the characterization of physical phenomena (ej. quality of electrical power systems, sensorial translation, etc...).
<br>

* **Non-stationary Signal Analysis**: While the Fourier Transform is primarily suited for stationary signals, its variations (like the *Short-Time Fourier Transform*) can hel analyze non-stationary signals by providing a time-frequency representation of the time series, revealing how the signals frequency content changes over time.


This method is crucial for time series analysis as its ability to bridge the time and frequency domains makes it a powerful tool across various fields. By addressing chalenges such as noise, non-stationarity, and feature extraction in time series analysis, the Fourier Transform amplifies the analytical toolkit available, enabling deeper insights into dynamic phenomenon. Some of its importan contributions in different fields include the following:

* **Frequency Analysis**: it enables the identification of dominant frequency components within a time series, helping understand periodicities, trends and noise.
<br>

* **Signal Filtering**: By identifying frequencies, it facilitates the process of filtering out target frequencies that may be considered irrelevant or noise in a specific context, enhancing target signal quality and specific information retrieval.
<br>

* **Data Compression**: In the frequency domain, it's possible to approximate the original signal with fewer components by identifying and carefully filtering frequencies that contribute most to the signal, and reverting back to the time domain (*inverse Fourier Transform*).
<br>

* **Feature Extraction**: In machine learning and data analysis, transforminng time series into the frequency domain can reveal features not readily available in the time domain (like those described above), improving model accuracy and general insights.

* **System Analysis**: In engineering and in physiscs, it helps in analyzing the response of systems to various frequencies (such as the above mentioned resonance frequencies), critical in the desing and analysis of diverse mechanical or electrical systems, acoustics and others.

#### Mathematical definition

To better understand the Fourier Transform we can break down the components of the mathematical equation:

$$\Large X(f) = \int_{-\infty}^{\infty}x(t)*e^{-j2\pi ft}dt$$

Where **X(f)** represents the *continuous Fourier Transform* of the signal ***x(t)***, giving us a complex function of frecuency ***f***. It describes the frequency components present in the original time domain dignal. The *real part* of ***X(f)*** corresponds to the **amplitude of the cosine components**, and the *imaginary part* corresponds to the **amplitude of the sine components** for each frequency ***f***.

* **x(t)**: is the original signal in the time domain, a function of time, representing how the signal varies over time.
<br>

* $\int_{-\infty}^{\infty}$: indicates that the transformation considers all values of **t** from negative to positive infinity, essentially summing up all the time based contributions across the entire duration of the signal.
<br>

* $e^{-j2\pi ft}$: is the *kernel* of the transformation, a complex exponential function where ***j*** is the *imaginary unit* ($\sqrt{-1}$) also known as ***i*** and $2\pi ft$ represents the **angular frequency** ($2\pi f$) multiplied by **time** (***t***), connecting the time domain signal with its frequency domain representation.
<br>

The exponential term $e^{-j2\pi ft}$ can be expanded into *Euler's formula* representation: $cos(2\pi ft) - jsin(2\pi ft)$. This shows that the Fourier Transform decomposes the signal into a sum of sinusoids, with frequencies rangin from negative to positive infinity, allowing us to analyze the signals frequency content. This **complex exponential** serves as a mathematical tool that allows us to project the time-domain signal into the frequency domain, separating it into its constituent frequencies.

Projection of the time-domain signal (***x(t)***) onto a basis of functions composed of sine and cosine waves of different frequencies is accomplished through the integral operation, which can be thought of as a continuous sum over all time points, combining the signal with a complex exponentiat of frequency ***f***. At each frequency ***f***, the product $x(t)e^{-j2\pi ft}$ mixes the signal with a sinusoid frequency. This operation effectively measures how well the signal alligns with, or is similar to, a sinusoid of frequency ***f***, **Decomposing the signal** into these frequencies. The result of this integration for each frequency ***f*** gives a *complex number*. The **magnitude or absolute value** of this number represents how strong the frequency component ***f*** is in the original signal ***x(t)***, or its **Amplitude**. The **argument or angle** of the complex number represents the **Phase shift** of this frequency component relative to the origin of time ***t = 0***. The **magnitude and angle** of a complex number is calculated as the magnitude and angle of its *vectorial representation*. For example:

$\large Z = 3 + 4j$

**Magnitude**:

$|Z| = \sqrt{3^2 + 4^2} = 5$

**Angle**: obtained by using the arctangent function.

$\theta = tan^{-1}\Big{(}\frac{\text{Real part}}{\text{Imaginary part}}\Big{)} = tan^{-1}(\frac{3}{4}) = 53.13Â°$ or 0.927 radians.

This means that the frequency component represented by Z = 3 + 4j has an amplitude of 5 and is phase-shifted by approximately 
53.13 degrees relative to the origin of time (t=0). The **magnitude** indicates the strength of this frequency component in the signal, while the **phase shift** indicates its timing or how much it is shifted from a reference point in its cycle at the beginning of the observation period.

By performing this operation for all frequencies ***f***, from negative to positive infinity, one constructs the full **frequency-domain representation *X(f)*** of the original time-domain signal. This spectrum ***X(f)*** reveals all the frequencies that make up ***x(t)***, along with their amplitudes and phase shifts.


### Discrete Fourier Transform (DFT)

The **Discrete Fourier Transform** is a mathematical technique used to transform a sequence of complex numbers, often representing a *discrete time series*, into a sequence of complex numbers representing its frequency domain components. The DFT is particularly useful in digital signal processing and time series analysis because it deals with discrete data, as opposed to the continuous signals handled by the Fourier Transform. Relevant differences between these two approaches include the following:

1. **Discrete vs Continuous data**: DFT operates on discrete data points (samples), whereas the continuous FT applies to continuous signals. The DFT assumes the signal is periodic and finite, repeating outside of the N points considered.
<br>

2. **Finite vs Infinite integration**: the FT analyzes signals over infinite range or a continuous interval of time, requiring integration. The DFT, in contrast, uses summation over a finite set of signal samples, reflecting the discrete nature of the data.
<br>

3. **Computational approach**: the DFT is particularly suited for computation, especially with the development of the ***Fast Fourier Transform***, which will be reviewed later, that efficiently computes the DFT of a sequence. The continuous FT, being an integral over an infinite range, is not directly computable but can be approximated for digital signals after sampling the original series and using the DFT or FFT.
<br>

4. **Output**: both the DFT and the FT produce spectra that represent the frequency content of the original signals. However, the DFT yields a discrete set of frequency components corresponding to the finite set of time-domain samples, whereas the continuous FT yields a constinuous spectrum.


#### Mathematical definition of the DFT

The DFT can be better understood by decomposing and explaining the components of its mathematical equation:

$$\Large X_k = \sum_{n=0}^{N}x_n * e^{-j2\pi nk/N}$$

for *k* = 0, 1, ..., N-1, where:

* $x_n$ represents the nth sample of the time-domain sequence. In the context of signal processing, ***xn*** could be the value of a signal at the nth time point, such as the amplitude of a sound wave, the brightness level of in an image pixel, or any discrete measurement taken at regular intervals.
<br>

* $X_k$: represents the kth component of the frequency domain. The DFT transforms each time-domain sample into a set of frequency-domain components, with $X_n$ indicating the contribution of the frequency *k* to the overall signal. Each $X_k$ is a complex number that encodes both the amplitude and phase of the frequency component *k*.
<br>

* *N*: is the total number of samples in the time-domain sequence. The DFT operates on a finite set of data points, and *N* denotes its size. The transformation assumes the input signal is periodic, with the period defined by *N*.
<br>

* $e^{-j2\pi nk/N}$ : is the kernel of the DFT, analogous to the exponential component of the continous FT seen above adapted for discrete samples. It's a complex exponential that oscilates, with the oscilation frequency determined by the product ***nk/N***. The kernel essentially represents the basis functions onto which the original signal ***xn*** is projected. The exponential terms include:

    * j, the imaginary unit
    
    * $2\pi$, ensuring the oscilation is periodic with a period of ***N***.
    
    * nk/N adjusts the frequency of the basis function for each compination of n (*the time index*) and k (*the frequency index*).
    
The DFT works by multiplying each time-domain sample by a ***complex exponential*** and summing these products over all ***n***. This process is repeated for each ***k***, from 0 to N-1, thus transforming the time-domain sequence into a frequency-domain representation. For each ***k***, the sum accumulates the contribution of all time-domain samples to the frequency component ***k***. The ***exponential*** acts as a **rotating vector or phasor** whose speed of rotation depends on ***k***. By multiplying ***xn*** by this rotating vector and summing, the DFT effectively measures how much of the signal's *energy* is present at the frequency corresponding to ***k***.

The result is a spectrum of complex numbers ***Xk***, each representing a different frequency component of the original signal. The *magnitude* of ***Xk*** (calculated as $\sqrt{Re(X_k)^2 + Im(X_k)^2}$) gives the *amplitude* of the frequency ***k***, while its *argument* (calculated as $tan^{-1}\Big{(}Im(X_k)/Re(X_k)\Big{)}$) gives the *phase shift* of that frequency relative to the start of the signal.

### Fast Fourier Transform (FFT)

The **Fast Fourier Transform** (FFT) is an algorithm designed to compute the DFT of a sequence much more efficiently than directly evaluating the DFT's definition. The FFT is not a different transformation from the DFT but rather a family of efficent algorithms to compute the DFT. The most common FFT algorithm is the **Cooley-Turkey algorithm**, but others like the **Radix-2, Bluestein's FFT** and the **split-radix FFT** also exist, each optimized for specific conditions or types of input.

While the DFT is widely applicable in theory, its direct computation is often impractical for large datasets due to computational costs. The DFT requires $O(N^2)$ complex multiplications, where N is the number of samples in the sequence. In contrast, the FFT significantly reduces this to ***O(N logN)***, making it much more efficient for a large N. This reduction in computational complexity is achieved, most famously through the **Cooley-Turkey algorithm**, by breaking down the DFT of a sequence of N points into smaller DFTs. This is possible when *N* is a power of 2, allowing the sequence to be divided repeatedly into smaller parts until they reach a manageable size (often until N=1 or N=2), where the DFT can be computed directly with minimal computation.

The key insight of the FFT is that the DFT of a sequence can be expressed in terms of the DFTs of two subsequences: one containing the even-indexed elements and the other containing the odd indexed elements. This decomposition leverages the periodicity and symmetry properties of the **complex exponentials** in the DFT formula, allowing the same computational work to be reused in calculating multiple ***Xk*** values.

For a Fourier Transform analysis, we'd typically look for frequency components in this time series to identify any dominant cycles. The next steps usually involve plotting the power spectrum to see which frequencies (or periods) stand out.

```python
# Function that will be used again for FFT visualization
def fft_visualization(signal, ylimit, plot =  True):
    # Perform the FFT on the original signal
    fft_values = np.fft.fft(signal)

    # Get the power spectrum (squared magnitude of the FFT values)
    power_spectrum = np.abs(fft_values)**2

    # Compute the corresponding frequencies
    frequencies = np.fft.fftfreq(len(fft_values), d=1) # sampling of 1

    # Anaalyze the peaks (ignoring zero frequency)
    positive_frequencies = frequencies[frequencies>0]
    positive_power = power_spectrum[frequencies>0]

    # Find the peak frequency and corresponding period
    peak_freq_index = np.argmax(positive_power) # get the index
    peak_freq = positive_frequencies[peak_freq_index] # get the corresponding frequency
    peak_period = 1/peak_freq if peak_freq!=0 else 0
    
    if plot:
        # Plot the power spectrum
        plt.figure(figsize=(15,7))
        plt.plot(np.abs(frequencies), power_spectrum, 
                 label=f"Peak frequency : {peak_freq:.4f}\nPeak period: {peak_period:.3f}")

        plt.ylim(-0.1,ylimit)
        plt.title("Power Spectrum of the Fourier Transform")
        plt.xlabel("Frequency (cycles per datapoint)")
        plt.ylabel("Power")
        plt.legend()
        plt.show()
    return positive_frequencies, positive_power
```

```python
# Perform the FFT on the cleaned temperature data
_,_=fft_visualization(df['Temp'].values, 1.2e8)
```

The above plot is a **Power spectrum** provided by the FFT which is a representation of the strength of the various frequency components (cycles per day) present in the time series. In the power spectrum, a prominent peak at a frequency suggests a regular cycle in the time series data. Here, the most significant frequency is at 0.00274 cycles per day, which corresponds to a peak period of about 365 days. This is a strong indicator of an annual cycle in the data, which makes sense for temperature data as it would reflect the seasons within each year. The peak at zero frequency is typically discarded because it represents the mean or DC component (i.e., a constant part of the signal, not a cycle), which does not contribute to the oscillations or trends we're usually interested in when analyzing time series data.

## 3. Complementary and Advanced Techniques

In time series analysis, data often exhibits non-stationary behavior. This could manifest as changes in the mean, variance, or periodic behavior throuought the series. Standard Fourier analysis may not be adequate for such data, as it assumes the time series is stationary throughout its length. To analyze non-stationary data, we need methods that can handle changes in the signal's frequency content over time.


### Short-Time Fourier Transform

The **Short-Time Fourier Transform** isan extension of the FT, designed specifically for analyzing signals whose frequency content changes over time. While FT gives the frequency content of the entire signal, it doesn't provide any information on when these frequencies occur. The STFT addressesthis limitation by providing a time-frequency representation of the signal.

Mathematically, the STFT is defined as follows:

$$\Large \text{STFT{x(t)}}(f,\tau) = \int_{-\infty}^{\infty}x(t)*w(t-\tau)*e^{-j2\pi ft}dt$$

Where:

* x(t) is the input signal as a function of time.
<br>

* $w(t-\tau)$ is a window function centered at time $\tau$. It is used to isolate a segment of the signal. The window function "slides" over the signal: for each position $\tau$, it zeroes out all but a small piece of the signal x(t), allowing the Fourier Transform to analyze just that piece. Common window functions include the ***rectangular, Hamming, Hanning*** and ***Gaussian windows***. 
<br>

* $\tau$ represents the time around which the window function is centered, and it's the parameter that "slides" along the input signal. As this parameter changes, the window slides over differentparts of the signal, allowing the STFT to anlyze the signal piece by piece. It's the variable that moves the window function along the time axis.
<br>

* $e^{-j2\pi ft}$ is the same complex exponential that projects the signal on the frequency domain, which tests the presence of frequency *f* in the windowed segment.
<br>

The result of this computation is a complex number for each frequency *f* and time $\tau$, representing the amplitde and phase of that frequency component at that time interval. By repeating this across different times $\tau$, we get a time-frequency representation of the original signal. This is typically displayed as a *spectrogram*, which shows how the energy of different frequencies varies over time.

Some further understading of the **window function** may be important before applying the STFT on our signal. As has been mentioed above, the window function is used to isolate a segment of the signal around a specific point in time. The purpose of the window function is to mitigate the effects of discontinuities at the edges of the segment when performing the FT on finite intervals. These mathematical functions are zero-valued outside of some interval. Typically, window functions are symmetric around the middle of the interval, approach a maximum in the middle, and taper away towards either extreme. Mathematically, when another function or data sequence is passed through a window function, the product is alse zero-valued outside the interval, and all that is left is the part where they overlap. 

Some common window functions are the following:

* **Rectangular Window**: mathematically, it's the simplest window: all samples within the window have equal weight and all samples outside are zero. Its simplicity has the benefit of not modifying the segment of the signal it is multiplying, but it does introduce significant **spectral leakage**, which will be further explained later, because the sudden drop to zero at the edges creates high frequency artifacts.
    
    The mathematical definition is the following:
    
    $w(n) = 1 \text{for } 0 â¤ n â¤ N - 1$ and $w(n) = 0$ otherwise.
<br>

* **Hann (Hanning) Window**: features a cosine shaped taper that smoothly decreases to zero at the edges of the window. This tapering reduces spectral leakage. The approach is good for general use where the balance between **frequency resolution** and *leakage* is required.
    
    The mathematical resolutio for this window is:
    
    $w(n) = 0.5\Big{(}1 - cos(\frac{2\pi n}{N-1})\Big{)}$ for 0 â¤ n â¤ N-1
<br>

* **Hamming Window**: The formulation is similar to the Hann window but with a different constant that leads to a slightly raised cosine shape, giving it a diferent taper at the edges. This provides a slightly better frequency resolution and less leakage into distant frequencies:

    $w(n) = \Big{(}0.54 - 0.46*cos(\frac{2\pi n}{N-1})\Big{)}$ for 0 â¤ n â¤ N-1
<br>    
  
* **Gaussian Window**: is based on the Gaussian distribution curve, providing excelent time and frequency resolution. Useful when the signal contains closely spaced frequencies or transient signals. The mathematical definition is the following:

    $w(n) = e^{-\frac{1}{2}(\frac{n - N/2}{\sigma N/2})}$ for 0 â¤ n â¤ N-1
    <br>
    
* **Kaiser window**: a family of windows with a shape controlled by a parameter $\beta$ that can be adjusted to trade off main lobe width (frequency resolution) and side lobe level (leakage). The mathematical definition is as follows:

    $w(n) = \frac{I_0\Big{(}\pi\beta\sqrt{1 - (\frac{2n}{N-1}-1)^2}\Big{)}}{I_0(\pi\beta}$
    
In all the above definitions **n** is the sample index and **N** is the total number of samples in the window. The window functions are all typically symetrical around $\frac{N-1}{2}$.

The window function when correctly identifyig target components we are looking for in a signal. Specifically, different windows have different benefits and trade-offs. As mentioned above, **spectral leakage** appears when the FT is applied to a finite signal and the edges can act like discontinuities. This is when the energy from the main frequency component *leaks* into adjacent frequencies. This is inherent to the nature of the Fourier Transform, which assumes that the signal is periodic and infinite. When the signal is truncated to fit within a finite window, the edges can create a step discontinuity, as if the wave suddenly stops and starts again, which is not a natural occurrence for most physical signals. The FT tries to represent these discontinuities at the window edges with high frequency components. By **tapering** (smoothing the ends of the signal), the window function reduces the effect of the sharp edges.

There is a trade-off between spectral leakage and **frequency resolution**, which refers to the ability to distinguish between two closely spaced frequencies in the spectrum. The frequency resolution is inversely related to the length of the time window: a longer window improves frequency resolution because it provides a more detailed view of the signal, allowing for a finer distinction between frequencies. 

In practice, this trade off choice depends on the specific requirements of the analysis. If precise frequency measurement of narrowband signals is needed, a window with high frequency resolution is chosen, accepting some degree of leakage. Conversely, if it's important to accurately measure the amplitude of frequency components without the inferferance of leakage, minimizing leakage is prefered, even though it may blur closely spaced frequency components.

The selection of each of these parameters depends on balancing between ***spectral leakage*** and ***frequency resolution*** when deciding over **window type**, ***frequency resolution*** and ***time resolution*** when decidint over **segment length**, and betweeen ***smoothness*** of time-frequency representation and the ***ability to capture changes*** in the signal when deciding on the **overlap**.

### Continuous Wavelet Transform

The **Continuous Wavelet Transform** is a time-frequency analysis technique that provides a representation of a signal in terms of scaled and translated versions of a mother wavelet.  Unlike the Fourier Transform, which breaks down a signal into sines and cosines that extend infinitely, the CWT uses wavelets that are localized in both time and frequency. Because of this feature, the CWT is able to analyze localized variations of power within a time series.

The CWT operates over two parameters: ***scale*** and ***position (translation)***. The following is a breakdown of the mathematical formulation of the CWT and its elements:

$$\Large CWT(s, \tau) = \frac{1}{\sqrt{s}}\int_{-\infty}^{+\infty}x(t)\psi^*\Big{(}\frac{t-\tau}{s}\Big{)}dt$$

In this formulation:

* x(t) is the original signal to be transformed.
<br>

* $\tau$ is the ***translational parameter***, which corresponds to the *shift* of the wavelet along the time axis. By translating the wavelet, we slide it along the time axis of the signal, allowint the wavelet to analyze different time intervals of the signal.
<br>

* s is the ***scale parameter***, which controls the dilation or compression of the wavelet. By scaling the mother wavelet, you change its width and the frequency band it analyzes. Small scale factors stretch the wavelet, making it narrower and better suited for analyzing high-frequency components of the signal. Large scale factors compress the wavelet, making it wider and better suited for analyzing low-frequency components.
<br>

* $\psi(t)$ is the ***mother wavelet***, which functions as a prototype for all other wavelet transformations used. The mother wavelet is a ***localized*** oscilating function both in frequency and time. Localization in time means that it is non-zero for a short duration. Localization in frequency means that it contains a band of frequencies centered around a particular frequency. It acts like a short-lived wave that is created and then quickly dies off. It is also ***compact***, in that it's non-zero values span over a finite interval, ensuring computational efficiency. Also, the wavelet should have a ***zero mean*** so that it can effectively represent fluctuations in the signal around the baseline. 

    A mother wavelet must satisfy an ***admissibility condition***, which ensures that the wavelet transform is invertible. This condition generally requires the wavelet to have a zero or near-zero value at the frequency of 0Hz (DC component). Under these condtions, there are many possible choices for the mother wavelet, each with different characteristics that make them suitable for different types of signals. Common examples include the ***Morlet wavelet*** (which looks like a sine wave modulated by a Gaussian envelope), the ***Daubechies wavelets*** (which are more abstract and optimized for particular properties), and the ***Mexican hat wavelet*** (which resembles the second derivative of a Gaussian). A mother wavelet $\psi(t)$ is chosen based on its similarity to the expected features in the signal.
    <br>

* $\psi^*(t)$ is the *complex conjugate* of the mother wavelet. As we've reviewed above, complex numbers are sometimes used to represent waves, it might be useful to enter into some detail on this subject here. 

    **Complex numbers tangent**: A complex wave can be thought of as a wave that has both a real and an imaginary part. These parts can be thought of as two sine waves that are out of phase with each other by 90 degrees. In the complex plane, this corresponds to one component having values along the *real* axis and the other along the *imaginary* axis. A **complex conjugate** of a complex number is created by changing the sign of the imaginary component. If we have a complex number *z* represented as *z = a + bi*, the complex conjugate would be $z^* = a - bi$. If we have a ***wave function*** that includes both a real and imaginary part, in the form of a ***complex exponential*** (like the one used in the FT and it's variants), represented by:
    
    $\psi(t) = Ae^{i(2\pi ft + \theta)}$
    
    The ***complex conjugate of the wave*** would be:
    
    $\psi^*(t) = Ae^{-i(2\pi ft + \theta)}$
    
    The effect of taking the complex conjugate in this context is to reverse the direction of the wave's rotation in the complex plane, effectively mirroring the wave across the real axis. 
    
    In the **CWT**, the *complex conjugate* of the **mother wavelet** is used during the transformation process. This is due to the ***inner product operation*** taken between ***x(t)*** and $\psi^*(t)$ that requires the conjugate when integrating over complex functions. The *inner product* is used to measure the similarity between the signal and the wavelet function at various scales and translations, quantifying how much of the *wavelet* is present in the signal at a specific time and scale. The use of the *complex conjugate* ensures that the resulting wavelet coefficients are also complex numbers, capturing both the *amplitude* (**magnitude**, how strong the particular frequency is) and the *phase* (**allignment**, or timming of the wavelet with respect to the signal) information of the signal's components at different scales and translations. By examining the magnitudes of the coefficients one can create a ***time-frequency representation*** of the signal, showing how its frequency content changes over time. 
<br>    

* $\frac{1}{\sqrt{s}}$ is a ***normalization*** factor to ensure the energy of the wavelet is independent of the **scale**. 

When implementing this equation and computing the CWT we must first choose a **mother wavelet** $\psi(t)$ based on its similarity to the expected features in the signal. The mother wavelet is **scaled** (*stretched* or *compressed*) by the **scale factor** *s*, and **translated** (*shifted*) along the time axis by the **translation factor** $\tau$. For each scale *s* and translation $\tau$, the **wavelet coefficient** $CWT(\tau,s)$ is calculated by taking the integral of the product of the signal *x(t)* and the **complex conjugate** of the scaled and translated mother wavelet $\psi^*(\frac{t-\tau}{s})$ . By varying *s* and $\tau$ over a range of values, a **matrix of wavelet coefficients** is formed, which represents the **time-frequency content** of the original signal.

The **magnitude** of each coefficient $|CWT(\tau,s)|$ indicates the similarity between the signal and the wavelet at a particular scale and translation. Large coefficients occur when the signal closely ressembles the shape of the wavelet.

We will now perform a CWT on the signal created using the **PyWavelets** library in python:

```python
# Wavelet Transform implementation:
import pywt

# Recap signal creation
xx = np.linspace(0, 30*np.pi, 1000)
y1 = np.sin(7*xx)
y2 = 2*np.sin(xx+0.5)
y3 = np.sin(xx)**2
y4 = np.sin(4*xx)
signal = np.zeros(xx.shape)

# Composite signal
signal = y1 + y2 
signal[:500]+= y3[:500] 
signal[500:]+= y4[500:]

# Define the range of scales for the CWT
scales = np.arange(1,128)

# Perform the CWT
coefficients, frequencies = pywt.cwt(data=signal,
                                     scales=scales,
                                     #The 'cmor' stands for Complex Morlet Wavelet
                                     #The parameter '1.0' indicates the bandwidt and '1.5' the center frequency of the wavelet
                                     wavelet='cmor1.0-1.5',
                                     sampling_period=1)

# Plot the CWT coefficients as a scalogram
plt.figure(figsize=(14,6))
plt.imshow(np.abs(coefficients), 
           extent=[0, len(signal), 1, max(scales)], 
           cmap='jet', 
           aspect='auto', 
           vmax=abs(coefficients).max(), 
           vmin=-abs(coefficients).max())
plt.colorbar(label='Magnitude')
plt.title("CWT Magnitude Spectrum of the signal")
plt.ylabel('Scales')
plt.xlabel('Time [Index]')
plt.show()
```

Before entering into the analysis of the plot created by the CWT, we should first explain some of the parameters selected in the code to understand what is being done here. The **scales** parameter represents *s* in the CWT equation reviewed above, and is an array that represents the range of values that *s* will take in the CWT calculation. A **scale** in wavelet analysis is inversely related to the frequency detected: smaller scales correspond to higher frequencies, and viceversa. The scale range of 1 to 128 provides a wide coverage that spans high to low frequencies, which is suitable for our signal that contains a broad spectrum of frequency content (y1 is seven times y2). After visualizing the first CWT implementation, the scale range can be adjusted depending on what we expect from the signal or if we want to adjust visual characteristics of the image generated (like resolution of the frequency component differences).

In the image above we can see somewhat clearly that there are 2 bands that span the whole timeline of the signal and 2 bands that span one the first half and the other the second. The former 2 would correspond to y1 and y2. Since y1 has a higher frequency than y2, and the scale is inversely related to the frequency content, the thick red band between scales 20 and 40 should be y1, and the thinner band near 115 should be y2. But, in this code, if we remove y1 from the signal the thin band at scale 115 is removed. This means that the deep red thick band is y2 and the above band is y1. The other 2 yellow thin bands correspond to the y3 and y4 components of the full signal. The effect of removing y1 can be seen here:

If we go back to the STFT made above, we can see that y2 has a higher power than y1, given by the amplitude in its definition *y2 = 2sin(x+0.5)*. Now that we have confirmed that the thicker band is y2, the amplitude of y2 is also represented in this CWT. But, given how the CWT works, there could be a case where a frequency with a lesser amplitude may show a higher magnitude. This can lead to some confusion, but it is explained by how the CWT operates and the magnitude of the components is generated. The magnitude of the components in a CWT scalogram can be influenced by several factors, not just the amplitude of the signal frequency components. The CWT magnigtude can be influenced by how well the energy of a wavelet at a particular scale matches the energy of the signal frequecy component it is anlyzing, a phenomenon known as **wavelet-energy matching**. ##### Complex Morlet Wavelet:

The above describes, in part, why **wavelet selection** is important, which brings us to another parameter in the code. Here, we've selected the **Complex Morlet wavelet** represented by the wavelet parameter ***"cmor"***. This wavelet is a popular choice for CWTs becaus of its balance between time and frequency localization. The Morlet wavelet is better suited for applications where a continuous frequency spectrum of the signal is of interest, such as in music signal analysis, EEG signal analysis, or any application where capturing the evolution of frequency content over time is crucial.

The *cmor wavelet* parameter also allows for selecting the **bandwidth**, which influences the number of oscilations within the wavelet shape. More specifically, it affects the spread of the frequency range that the wavelet covers. A higher bandwidth parameter means the wavelet function will cover a wider range of frequencies. Increasing the **bandwidth** parameter increases the *frequency range* that the wavelet covers. This parameter influences the wavelet's "width" in the frequency domain, affecting its ability to resolve different frequencies. Here, we've selected a *1.0* bandwidth, which has correctly identified the 4 frequency domain components of the signal. We can choose other values for this parameter to see the effect they have. What should be expected is that a higher badnwidth would visualize more frequencies (if present) and a lower would identify less.

Another parameter defined in the *cmor wavelet* is the **center frequency parameter** given here by *1.5*. This parameter refers to the frequency around which the wavelet is centered in the frequency domain. For the cmor wavelet, defined by a complex exponential (sine wave) modulated by a Gaussian envelope, the center frequency parameter determines the sine wave's frequency. To better illustrate this here is the mathematical formulation of the cmor wavelet: 

$\large\psi(t) = \pi^{-1/4}e^{j2\pi f_ct}e^{-t^2/2}$ 

where $f_c$ is the **center frequency parameter** selected. The center frequency allows the wavelet to be tuned to specific frequencies of interest within a signal. By adjusting the center frequency, you can focus the wavelet transform on analyzing how the amplitude of these frequencies changes over time, giving the wavelet its *frequency sensitivity*. 

The wavelets **time localization** is given by the **Gaussian envelope**: $e^{-t^2/2}$, which confines the wavelet in time. This ensures the wavelet has a compact support and can analyze localized features of the signal. $\pi^{-1/4}$ is a **normalization factor** that ensures the wavelet has a unit energy, making the analysis consistent and independent of the wavelet's scale.

When implementing the **Complex Morlet wavelet** in the CWT calculation, the first step is ***scaling and translating*** the wavelet. For each **scale** *s* and **translation** $\tau$ value, the wavelet is scaled and translated to match the different parts of the signal. The scaled and translated wavelet function is given by:

$\large\psi_{s,\tau}(t) = \frac{1}{\sqrt{s}}\psi\Big{(}\frac{t-\tau}{s}\Big{)}$

As explained above, *s* here is inversely related to the frequency and, given the equation, we can better understand why. Lower scales (smaller s) compress the wavelet, raising its frequency, making it more sensitive to higher frequencies in the signal being analyzed. This is because the compressed wavelet oscillates more rapidly, aligning better with higher frequency components of the signal. $\tau$ is the translation factor, determining the position of the wavelet along the signal's timeline.

The **wavelet coefficient** at each scale *s* and each translation $\tau$ is calculated by taking the **inner product** of the signal *x(t)* with the conjugate of the scaled and translated wavelet as seen here:

$\large CWT(s,\tau) = \int x(t) \psi^*_{s,\tau}(t)dt$

By repeating the coefficient calculation for a range of scales and translations, a **matrix of coefficients** is constructed. The matrix represents the signal's time-frequency content, where each coefficient's magnitude reveals the presence and strength of signal components corresponding to the wavelet's frequency at each moment in time. With the coefficient matrix the **scalogram**, a visual representation of the CWT, can be plotted as seen above.

##### Mexican Hat wavelet

The MHW, also known as the **Ricker wavelet**, is mathematically defined as the second derivative of a *Gaussian function*:

$\large\psi(t) = (1 - t^2)e^{-t^2/2}$

This wavelet is particularly useful for detecting and characterizing localized features in signals, such as spikes, bumps, and ridges, due to its shape. This characteristic shape makes it adept at highlighting rapid changes in signal intensity, which is why it's often used in image processing for edge detection and in geophysics for seismic analysis.

The following is a visual representation of this wavelet:

##### Daubechies Wavelet:

The **Daubechies wavelets** are a family of orthogonal wavelets defining discrete wavelet transforms and characterized by a maximal number of vanishing moments for some given support width. These wavelets are used in signal and image processing due to their compact cupport and ability to represent data with a small number of coefficients efficiently.

Daubechies wavelets are extensively used in signal compression because of their ability to compactly represent signals and images without significant loss of information. They are also used in denoising applications where the goal is to remove noise from the signal while preserving. Because of their ability to extract spatial information, this approach is useful in image feature extraction (ej. finger print recognition). Also, in scientific computing, Daubechies wavelets serve as a basis for solving partial equations numerically. Their properties allow for sparse representations of the solution space, making computations more efficient.

Unlike wavelets like the Mexican Hat or Morlet, which have explicit formulas in the time domain, Daubechies wavelets are defined by their properties in the frequency domain and through recursion relations in the time domain, making their explicit mathematical definition more complex. Mathematically, these wavelets are constructed to satisfy both the **scaling equation**:

$\large\phi(x) = \sqrt{2}\sum_{k=0}^{2N-1}h_k\psi(2x - k)$

and the **wavelet function** through a similar relation:

$\large\phi(x) = \sqrt{2}\sum_{k=0}^{2N-1}(-1)^kh_{2N-1-k}\psi(2x - k)$

where $h_k$ are the coefficients of a low-pass filter, which are derived from the requirements of orthogonality and compact support. *N* is the order of the Daubechies wavelet, which also indicates the number of vanishing moments. 

While there's no simple closed-form expression for Daubechies wavelets in the time domain, their most defining characteristics are the ***filter coefficient*** $h_k$, which are calculated through an iterative process designed to maximize the smoothness of the wavelet for a given order *N*. These coefficients can be found in tables or calculated using specialized software packages.

```python
wavelet = 'db2'  # Daubechies wavelet of order 2 (db2)
phi, psi, x = pywt.Wavelet(wavelet).wavefun(level=5)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, phi)
plt.title('Scaling Function (Phi)')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, psi)
plt.title('Wavelet Function (Psi)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

This example plots the ***scaling function*** $\phi$ and ***wavelet function*** $\psi$ of the Daubechies wavelet of order 2 (*db2*) providing a visual representation of these functions.

## 4. Forecasting using the Fourier Transform

The following is the same signal created above, but now repeated 3 times:

```python
# Recap signal creation
xx = np.linspace(0, 90*np.pi, 3000)
y1 = np.sin(7*xx)
y2 = 2*np.sin(xx+0.5)
y3 = np.sin(xx)**2
y4 = np.sin(4*xx)
signal = np.zeros(xx.shape)

# Composite signal
signal[:1000] = y1[:1000] + y2[:1000] 
signal[:500]+= y3[:500] 
signal[500:1000]+= y4[500:1000]
signal[1000:2000] += signal[:1000]
signal[2000:3000] += signal[:1000]

plt.figure(figsize=(14,6))
plt.plot(xx, signal)
plt.show()
```

```python
# Fourier Transform
fft_result = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(signal), d=(90*np.pi/3000))

# Plot the magnitude of the Fourier Transform
plt.figure(figsize=(10, 4))
plt.plot(frequencies, np.abs(fft_result))
plt.title('Fourier Transform of the Signal')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(0, 0.8)  # Focus on positive frequencies
plt.show()
```

```python
# Identify dominant frequencies (simplified as an example)
dominant_freqs = frequencies[(np.abs(fft_result) > np.percentile(np.abs(fft_result), 99)) & (frequencies > 0)]

# Extend the time for forecasting
extended_time = np.linspace(0, 135*np.pi, 4500)

# Initialize an extended signal array for the reconstructed signal
reconstructed_signal = np.zeros_like(extended_time)

# Reconstruct signal using dominant frequencies (simplified approach)
for freq in dominant_freqs:
    index = np.where(frequencies == freq)[0][0]
    amplitude = np.abs(fft_result[index])/1500
    phase = np.angle(fft_result[index])
    reconstructed_signal += amplitude * np.cos(2 * np.pi * freq * extended_time + phase)

# Plot original and forecasted signal
plt.figure(figsize=(14, 6))
plt.plot(xx, signal, label='Original Signal', color='blue')
plt.plot(extended_time[3000:], reconstructed_signal[3000:], label='Forecasted Signal', linestyle='--', color='red')
plt.title('Signal Forecasting Using Fourier Analysis')
plt.xlabel('Time')
plt.ylabel('Signal')
plt.legend()
plt.grid(True)
plt.show()
```

## 5. Comparing the Fourier Transform 

Time series analysis involves various methods, each with its advantages and aplications, depending on the nature of the data and the anlysis goals. The **Fourier Transform** decomposes a signal into its constituent frequencies, providing a **frequency-domain** representation. This method excels in analyzing periodic components within the time series, and is excellent for identifying and isolating periodic patterns and trends within the data. The FT allows for the reconstruction of the signal with the option to filter out noise or unwanted frequencies. Also, this approach can handle **non-stationay** data where the frequency components vary over time, a feature that is not available in other **time-domain** approaches that include ARIMA models and Exponential smoothing.

Frequency-domain results can be less intuitive to interpret compared to these time-domain methods. Moreover, forecasting using the FT is not as straightforward because it isn't a **model-based** method, requiring additional steps to predict future values. 

**AutoRegressive Integrated Moving Average (ARIMA)** models time series based on *past values* (auto-regressive), *difference of past values* (integrated), and *past forecast errors* (moving average). It's highly versatile for forecasting and certain non-stationary time series. ARIMA provides a model-based forecasting method, this model explicitly forecast future values based on historical data. ARIMAs flexibility allows it to model a wide range of time series with various levels of complexity, and allows fot hypothesis testing and confidence intervals for forecasts. 

ARIMA models require a stationary time series, this often involves transforming the data to be stationary, which might not always be straightforward or intuitive. Also, **parameter selection** (choosing the right order of p, d, q) can be complex and usually requires expertise or automated selection methods.

The choice between these methods depends on the specific characteristics of the time series, the analysis objectives, and the desired balance between model complexity, interpretability, and computational efficiency.
